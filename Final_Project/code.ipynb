```python
import re


# Prepare the text for analysis by converting it to uppercase and splitting into words
def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = text.upper()  # Convert to uppercase
    words = text.split()  # Split into words
    return words

# Perform sentiment analysis on the text
def sentiment_analysis(words, emotional_dict):
    positive_score = 0
    negative_score = 0
    uncertainty_score = 0
    litigious_score = 0
    strong_modal_score = 0
    weak_modal_score = 0
    constraining_score = 0
    complexity_score = 0
    for word in words:
        if word in emotional_dict.index:
            if emotional_dict.at[word, 'Positive'] > 0:
                positive_score += 1
            if emotional_dict.at[word, 'Negative'] > 0:
                negative_score += 1
            if emotional_dict.at[word, 'Uncertainty'] > 0:
                uncertainty_score += 1
            if emotional_dict.at[word, 'Litigious'] > 0:
                litigious_score += 1
            if emotional_dict.at[word, 'Strong_Modal'] > 0:
                strong_modal_score += 1
            if emotional_dict.at[word, 'Weak_Modal'] > 0:
                weak_modal_score += 1
            if emotional_dict.at[word, 'Constraining'] > 0:
                constraining_score += 1
            if emotional_dict.at[word, 'Complexity'] > 0:
                complexity_score += 1

    # Calculate the net sentiment score
    net_score = positive_score - negative_score
    return net_score, positive_score, negative_score, uncertainty_score, litigious_score, strong_modal_score, weak_modal_score, constraining_score, complexity_score 

# Your text data
with open('Kaggle_NASDAQ_10_2016To2020/Transcripts/AAPL/2016-Jan-26-AAPL.txt', 'r') as file:
    text_data = file.read()

# Preprocess the text and split into words
words = preprocess_text(text_data)

# Run sentiment analysis
net_score, positive_score, negative_score, uncertainty_score, litigious_score, strong_modal_score, weak_modal_score, constraining_score, complexity_score = sentiment_analysis(words, emotional_dict)

print(f'Net Sentiment Score: {net_score}')
print(f'Positive Score: {positive_score}')
print(f'Negative Score: {negative_score}')
print(f'Uncertainty Score: {uncertainty_score}')
print(f'Litigious Score: {litigious_score}')
print(f'Strong Modal Score: {strong_modal_score}')
print(f'Weak Modal Score: {weak_modal_score}')
print(f'Constraining Score: {constraining_score}')
print(f'Complexity Score: {complexity_score}')
```
