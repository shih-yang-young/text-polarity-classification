```python
companies_folders = [
    'Kaggle_NASDAQ_10_2016To2020/Transcripts/AAPL',
    'Kaggle_NASDAQ_10_2016To2020/Transcripts/AMD',
    'Kaggle_NASDAQ_10_2016To2020/Transcripts/AMZN',
    'Kaggle_NASDAQ_10_2016To2020/Transcripts/ASML',
    'Kaggle_NASDAQ_10_2016To2020/Transcripts/CSCO',
    'Kaggle_NASDAQ_10_2016To2020/Transcripts/GOOGL',
    'Kaggle_NASDAQ_10_2016To2020/Transcripts/INTC',
    'Kaggle_NASDAQ_10_2016To2020/Transcripts/MSFT',
    'Kaggle_NASDAQ_10_2016To2020/Transcripts/MU',
    'Kaggle_NASDAQ_10_2016To2020/Transcripts/NVDA',
]
```

```python
month_dict = {
    'Jan': '01',
    'Feb': '02',
    'Mar': '03',
    'Apr': '04',
    'May': '05',
    'Jun': '06',
    'Jul': '07',
    'Aug': '08',
    'Sep': '09',
    'Oct': '10',
    'Nov': '11',
    'Dec': '12'
}
def extract_date(data):
    parts = data.split('-')
    year = parts[0]
    month = month_dict[parts[1]]   
    day = parts[2]
    return f"{year}-{month}-{day}"
```

```python
import os
import pandas as pd

# Initialize an empty DataFrame
transcripts_df = pd.DataFrame(columns=['Company', 'Date', 'Content'])

# Iterate over each company folder
for company_folder in companies_folders:
    # Get the company name from the folder path
    company_name = company_folder.split('/')[-1]
    # Iterate over each file in the company's folder
    for filename in os.listdir(company_folder):
        if filename.endswith('.txt'):
            file_path = os.path.join(company_folder, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
            transcripts_df = transcripts_df._append({
                'Company': company_name, 
                'Date': extract_date(filename), 
                'Content': content
            }, ignore_index=True)

transcripts_df
```
