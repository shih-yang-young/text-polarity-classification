```python
import pandas as pd
amazon_review_sa_binary_csv_amazon_train = pd.read_csv('amazon_review_sa_binary_csv/train.csv')
amazon_review_sa_binary_csv_amazon_test = pd.read_csv('amazon_review_sa_binary_csv/test.csv')
```

```python
amazon_train = amazon_review_sa_binary_csv_amazon_train.copy()
amazon_train['class_index'] = amazon_train['class_index'].replace({2: 1, 1: 0})
amazon_train = amazon_train.rename(columns={'class_index': 'LABEL', 'review_text': 'TEXT'})
amazon_train['TEXT'] = amazon_train['review_title'] + " " + amazon_train['TEXT']
amazon_train = amazon_train.drop(columns=['review_title'])
amazon_train
```

```python
amazon_test = amazon_review_sa_binary_csv_amazon_test.copy()
amazon_test['class_index'] = amazon_test['class_index'].replace({2: 1, 1: 0})
amazon_test = amazon_test.rename(columns={'class_index': 'LABEL', 'review_text': 'TEXT'})
amazon_test['TEXT'] = amazon_test['review_title'] + " " + amazon_test['TEXT']
amazon_test = amazon_test.drop(columns=['review_title'])
amazon_test
```

```python
amazon_data = pd.concat([amazon_train, amazon_test], axis=0)
amazon_data = amazon_data.reset_index(drop=True)
amazon_data
```

```python
amazon_short_text_data = amazon_data[amazon_data['TEXT'].str.len() <= 250]
amazon_short_text_data = amazon_short_text_data.reset_index(drop=True)
amazon_short_text_data = amazon_short_text_data[amazon_short_text_data['TEXT'].str.match(r'^[A-Za-z0-9 .,!?\'"-]+$')]
amazon_short_text_data = amazon_short_text_data.reset_index(drop=True)
amazon_short_text_data
```

```python
label_counts = amazon_short_text_data['LABEL'].value_counts()

# 顯示計數結果
print(label_counts)
```

```python
amazon_short_text_data['row_id'] = amazon_short_text_data.index
amazon_short_text_data
```

```python
amazon_short_text_data.to_csv('amazon_short_text_data.csv', index=False)
```
